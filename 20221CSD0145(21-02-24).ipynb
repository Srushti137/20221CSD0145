```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step 1")
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,500,1000)
print(x)
ax.plot(x, np.sin(x));
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x, np.cos(x));
ax.plot(x, np.sin(x));
ax.plot(x, np.tan(x));
ax.set_title('Simple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x, np.cos(x), label = 'cos');
ax.plot(x, np.sin(x), label = 'sin');
ax.plot(x, np.tan(x));
ax.set_title('Multiple Lines')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x, np.cos(x), label = 'cos', color = 'black');
ax.plot(x, np.sin(x), label = 'sin', color = 'red');
ax.plot(x, np.tan(x));
ax.set_title('Multiple Lines')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x, np.cos(x), label = 'cos', linestyle = 'dotted');
ax.plot(x, np.sin(x), label = 'sin', linestyle = 'dashed');
ax.plot(x, np.sin(x+1), label = 'sin x+1', linestyle = 'dashdot');
ax.set_title('Multiple Lines')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```
